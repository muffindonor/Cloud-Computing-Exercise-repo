```python
import matplotlib.pyplot as plt
import json

dictionary = json.load(open('/3_1.json','r'))
xAxis = [key for key, value in dictionary.items()]
yAxis = [value for key, value in dictionary.items()]
plt.grid(True)

##LINE GRAPH
plt.plot(xAxis,yAxis,color='#8a5082', marker='o')
plt.xlabel('variable')
plt.ylabel('value')
plt.show()
```
